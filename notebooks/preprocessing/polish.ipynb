# **Polish Dataset Preprocessing**

```python
utilities.create_directory(polish_prep_dataset_path)
```

```python
data1 = utilities.read_csv(polish_dataset_path + '1year.csv', header=None)
data2 = utilities.read_csv(polish_dataset_path + '2year.csv', header=None)
data3 = utilities.read_csv(polish_dataset_path + '3year.csv', header=None)
data4 = utilities.read_csv(polish_dataset_path + '4year.csv', header=None)
data5 = utilities.read_csv(polish_dataset_path + '5year.csv', header=None)
data = pd.concat([data1, data2, data3, data4, data5])
data
```

#### Handle special missing values.

```python
utilities.replace_to_nan(data, "?")
```

#### Fix column dtypes.

```python
utilities.object_to_float64(data)
data['defaulted'] = data['defaulted'].astype(bool)
```

#### Plot features data types.

```python
plotting.plot_dtypes_hist(data, '', 'Dtypes', 'Counts',
                          save_path=polish_prep_dataset_path + 'dtypes.pdf')
```

#### Data cleaning.

```python
before_data_cleaning_count = len(data)
```

#### Plot class distribution.

```python
plotting.plot_hist(['Defaulted', 'Non-Defaulted'], data.defaulted.value_counts().values,
                   '', 'Classes', 'Samples Count',
                   save_path=polish_prep_dataset_path + 'class_distribution.pdf')
```

#### Plot missing values histogram.

```python
labels = [str(idx) for idx in data.isna().sum(axis=0).index]
values = data.isna().sum(axis=0).values
plotting.plot_hist(labels, values, '', 'Features', 'Missing values', figsize=(22,8),
                   rotated_ticks=True, save_path=polish_prep_dataset_path + 'missing_values-1.pdf')
```

#### Apply optimal binninf and WoE.

```python
data, features_scores = preprocessing.opt_bin_woe(data=data, solver="cp",
                                                  outlier_detector="range",
                                                  save_path=polish_prep_dataset_path,
                                                  verbose=False)
```
